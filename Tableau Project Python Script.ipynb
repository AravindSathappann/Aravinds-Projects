```python
import requests
import json
import pandas as pd

url = 'https://www.dunkindonuts.com/bin/servlet/dsl'

response = requests.post(
    url,
    headers={"User-Agent": "Mozilla/5.0"},
    data={
        'service': "DSL",
        'origin': '37.7749, -122.4194',
        'radius': "100000",
        'maxMatches': "100000"
    }
)

content = response.json()

stores = content["data"]["storeAttributes"]

df = pd.json_normalize(stores)

df.to_csv("dunkinDonuts.csv", index=False)

print("CSV created successfully.")
```
