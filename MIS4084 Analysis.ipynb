```python
import requests
import json
import pandas as pd

url = 'https://www.dunkindonuts.com/bin/servlet/dsl'
response = requests.post(url, allow_redirects=False, data={
    'service': "DSL",
    'origin': '37.7749, -122.4194',
    'radius': "100000",
    'maxMatches': "100000"
})

content = json.loads(response.content)

stores = content["data"]["storeAttributes"]

# Convert geoJson field
for store in stores:
    store["geoJson"] = json.loads(store["geoJson"])

df = pd.json_normalize(stores)

df["longitude"] = df["geoJson"].apply(lambda x: x["coordinates"][0])
df["latitude"] = df["geoJson"].apply(lambda x: x["coordinates"][1])

df.to_csv("dunkinDonuts.csv", index=False)

print("CSV created successfully.")
```

```python
import requests
import json
import pandas as pd

url = 'https://www.dunkindonuts.com/bin/servlet/dsl'

response = requests.post(
    url,
    headers={"User-Agent": "Mozilla/5.0"},
    data={
        'service': "DSL",
        'origin': '37.7749, -122.4194',
        'radius': "100000",
        'maxMatches': "100000"
    }
)

content = response.json()

stores = content["data"]["storeAttributes"]

df = pd.json_normalize(stores)

df.to_csv("dunkinDonuts.csv", index=False)

print("CSV created successfully.")
```
